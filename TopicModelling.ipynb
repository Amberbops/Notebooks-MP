## **Step3: Some Necessary Preprocessing**

```python
stop_set=set(stopwords.words('english'))
exclude_set=set(string.punctuation)
lemmatize=WordNetLemmatizer()
def clen_doc(doc):
  stop_free=" ".join([i for i in doc.lower().split() if i not in stop_set])
  punc_free=" ".join([i for i in stop_free.split() if i not in exclude_set])
  normalized=" ".join(lemmatize.lemmatize(w) for w in punc_free.split())
  return normalized
doc_clean=[clen_doc(doc).split() for doc in doc_complete]
```

##*Step 4: Create LDA model using gensim*

```python
dictionary = corpora.Dictionary(doc_clean)
doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_clean]
Lda=gensim.models.ldamodel.LdaModel
ldamodel=Lda(doc_term_matrix, num_topics=5, id2word=dictionary, passes=300)
topics=ldamodel.print_topics(num_topics=5, num_words=5)
for topic in topics:
  print(topic)
```

## **Conditional Random Fields**
Conditional Random Fields (CRF) is a sequence modelling technique which is most commonly used for pattern matching using structured prediction.

Generally in other models we do stemming and lemmatization or get vocabulary which makes our data lose its sequentiality or pattern. Which results in a loss of information or context.  CRF tells us that by having a sequential pattern i.e. having the labels of the nearby data(could be tokens or images) a lot of information is taken into account for predicting the labels on the desired column of data.

CRF uses parts of speech of the the tokens created from our text to learn the patterns of how each of these tokens occur in different contexts. It can then predict the labels on new data.

For example, given the below sentences tagged with metadata - account number:

```python
def convert_document_to_feature_functions(document, i):
    # Building features for feature functions f(s, i , li-1 , l i)
    word = document[i][0]

    features={
        'Currword': word,
    }
    # Features from previous word
    if i > 0:
        prevword = document[i-1][0]
        features['Prevword'] = prevword
    else:
        features['BOS'] = True # Special "Beginning of Sequence" tag

    # Features from next word
    if i < len(document)-1:
        nextword = document[i+1][0]
        features['Nextword'] = nextword
    else:
        features['EOS'] = True # Special "End of Sequence" tag
    return features
```
